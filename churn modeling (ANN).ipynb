```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Convert categorical variables into dummy/indicator variables
df = pd.get_dummies(dataset, columns=['Geography', 'Gender'], drop_first=True)
```

```python
# Split the dataset into features (X) and target variable (y)
X = df.drop(columns=['Exited','Surname'])
y = df['Exited'].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=46)
```

```python
# Feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Build the ANN model
classifier = Sequential()
classifier.add(Dense(3, kernel_initializer='uniform', activation='relu', input_dim=X_train.shape[1]))
classifier.add(Dense(3, kernel_initializer='uniform', activation='relu'))
classifier.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
classifier.add(Dense(3, kernel_initializer='uniform', activation='relu'))
classifier.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
classifier.add(Dense(3, kernel_initializer='uniform', activation='relu'))
classifier.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
classifier.fit(X_train, y_train, batch_size=20, epochs=100)
# Evaluate the model on the test set
loss, accuracy = classifier.evaluate(X_train, y_train)
print(f'Test Loss: {loss}')
print(f'Test Accuracy: {accuracy}')
```
