Now showing the scatter_matrix of the dataset using the scatter_matrix() function.

```python
pd.plotting.scatter_matrix(dataset,figsize=(15,10))
plt.title('scatter plot')
plt.show()
```

Now showing the pearson correlation matrix of the dataset. First correlation is found out using the corr() function and then it is plotted using the using pyplot of the matplotlib

```python
correlations = dataset.corr(method='pearson')
names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age','Outcome']
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
